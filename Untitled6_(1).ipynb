```python
import sys
import pdb
import pprint
import logging
import os
import random

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils import data
import numpy as np
import tqdm.auto as tqdm
from pathlib import Path
from argparse import Namespace
from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
```

```python
import pandas as pd

# 指定Excel文件路径
excel_file = "/content/data/data.xlsx"

# 读取Excel文件中的第一个工作表
df = pd.read_excel(excel_file)

names = df["target_name"].tolist()
names = list(set(names))
selected_columns = ['x', 'y', 'ax', 'ay', 'vx', 'vy']
time_steps = 10  # 输入序列的时间步长
pred_steps = 5  # 预测的时间步长
X = []
y = []
```

```python
for i in range(len(names)):
    one_track = df.loc[df["target_name"]==names[i]]
    one_track = one_track[selected_columns].values
    for i in range(len(one_track) - time_steps - pred_steps + 1):
        X.append(one_track[i: i + time_steps])
        y.append(one_track[i + time_steps: i + time_steps + pred_steps])

X = np.array(X)
y = np.array(y)

# np.random.seed(1)
# np.random.shuffle(X)
# np.random.seed(1)
# np.random.shuffle(y)


train_size = int(0.8 * len(X))
train_X, test_X = X[:train_size], X[train_size:]
train_y, test_y = y[:train_size], y[train_size:]
```

```python
# 归一化数据
scaler = MinMaxScaler()
scaler.fit(train_X.reshape(-1,train_X.shape[2]))

for i in range(len(train_X)):
    train_X[i] = scaler.transform(train_X[i])
    train_y[i] = scaler.transform(train_y[i])

for i in range(len(test_X)):
    test_X[i] = scaler.transform(test_X[i])
    test_y[i] = scaler.transform(test_y[i])

train_X = train_X.reshape(train_X.shape[0],train_X.shape[2],train_X.shape[1])
train_y = train_y.reshape(train_y.shape[0],train_y.shape[2],train_y.shape[1])
test_X = test_X.reshape(test_X.shape[0],test_X.shape[2],test_X.shape[1])
test_y = test_y.reshape(test_y.shape[0],test_y.shape[2],test_y.shape[1])
```

```python
print(train_X.shape)
print(test_X.shape)
```

```python
train_X = torch.from_numpy(train_X).float()
train_y = torch.from_numpy(train_y).float()
test_X = torch.from_numpy(test_X).float()
print(train_X.shape)
```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        out, _ = self.lstm(x, (h0, c0))
        print(out.shape)
        out = self.fc(out)

        return out



# 定义模型的参数
input_size = 10
hidden_size = 64
num_layers = 2
output_size = pred_steps
```

```python
model = LSTMModel(input_size, hidden_size, num_layers, output_size)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.003)

valid_losses = []
train_losses = []
num_epochs = 100
for epoch in range(num_epochs):
    # 前向传播和计算损失
    outputs = model(train_X)
    print(train_y.shape)
    loss = criterion(outputs, train_y)
    train_losses.append(loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # 每训练10个epoch打印一次损失
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

plt.plot(range(1, num_epochs+1), train_losses, label='Training Loss')

plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch.optim as optim
criterion = nn.MSELoss()
# 定义优化器
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)  # 调整学习率为 0.01，动量为 0.9

# 训练循环
for epoch in range(num_epochs):
    # 前向传播和计算损失
    outputs = model(train_X)
    loss = criterion(outputs, train_y)
    train_losses.append(loss.item())

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # 每训练10个epoch打印一次损失
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')


plt.plot(range(1, num_epochs+1), train_losses, label='Training Loss')

plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
model.eval()  # 设置模型为评估模式
with torch.no_grad():
    predicted = model(test_X)


# 打印预测结果
print(predicted)
```
